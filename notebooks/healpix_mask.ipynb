```python
import matplotlib.pyplot as plt
import numpy as np

import healpy as hp
```

```python
NSIDE = 128
print(
    "Approximate resolution at NSIDE {} is {:.2} deg".format(
        NSIDE, hp.nside2resol(NSIDE, arcmin=True) / 60
    )
)
```

```python
NPIX = hp.nside2npix(NSIDE)
print(NPIX)
```

```python
theta, phi = hp.pix2ang(NSIDE, np.arange(NPIX))
ra = np.degrees(phi)
dec = np.degrees(np.pi / 2 - theta)
```

```python
from cmass.survey.tools import BOSS_angular

wdir='/automnt/data80/mattho/cmass-ili'
mask = BOSS_angular(ra, dec, wdir=wdir)
```

```python
hp.mollview(mask, title='BOSS DR12 CMASS-NGC mask')
hp.graticule()
```

```python
hp.write_map("CMASS_NGC_DR12_mask.fits", mask, overwrite=True)
```
