```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
wdir = '/home/mattho/data/quijote/density_field/latin_hypercube_HR/0'
grids = [128, 64, 32, 16, 8, 4]

dfs = [np.load(f'{wdir}/dfmatt_m_{grid}_z=0.npy')
       for grid in grids]
```

```python
f, axs = plt.subplots(2, 3, figsize=(10, 6))
axs = axs.flatten()

w=1
for i, df in enumerate(reversed(dfs)):
    x = np.mean(df[:w], axis=0)
    axs[i].imshow(x)
    axs[i].set_title(f'N = {grids[-i-1]}^3')
    axs[i].axis('off')
    w*=2
```
